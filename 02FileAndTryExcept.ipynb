```
with open("/content/sample_data/email.txt", "r") as file:
  alphabatically_sort = file.readlines()
  alphabatically_sort = [line.lower() for line in alphabatically_sort]
  alphabatically_sort.sort()
  unique_lines = set(alphabatically_sort)
  content_of_file = file.read()

with open("/content/sample_data/email.txt", "r") as file:
  content_of_file = file.read()

content_of_file_to_lower = content_of_file.lower()
print(content_of_file_to_lower)
print(unique_lines)

with open("/content/sample_data/email.txt", "w") as file:
  file.writelines(unique_lines)
```
