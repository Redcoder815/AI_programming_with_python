```
def reshape_array(arr, sp):
  numpy_array = np.array(arr)

  if numpy_array.size != sp[0] * sp[1]:
    return []

  numpy_array = numpy_array.reshape(sp[0], sp[1])
  return numpy_array.tolist()
```

```
print(reshape_array([1,2,3,4,5,6,7,8], (2,4)))
```

```
def scaler_multiply(arr, scaler):
  num_array = np.array(arr)
  return num_array * scaler
```

```
print(scaler_multiply([1,2,3], 2))
```

```
def digonal(x):
  return np.diag(x)
```

```
print(digonal([1,2,3]))
```

```
def elementwise_sum(a,b):
  a = np.array(a)
  b = np.array(b)
  if a.size != b.size:
    return -1
  return a + b
```

```
print(elementwise_sum([1,2,3], [4,5,6]))
```

```
def dot_product(a,b):
  a = np.array(a)
  b = np.array(b)

  return np.dot(a,b)
```

```
print(dot_product([1,2,3], [4,5,6]))
```
