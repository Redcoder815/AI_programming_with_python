```
import numpy as np
import pandas as pd

np.random.seed(24)

num_employees = 50  # you can increase this if needed

# Employee data
emp_ids = np.arange(101, 101 + num_employees)
emp_ages = np.random.randint(20, 55, num_employees)
emp_income = np.random.randint(35000, 95000, num_employees)

# Fixed set of roles (repetition allowed)
roles_list = [
    "HR Manager",
    "Software Developer",
    "Accountant",
    "Marketing Executive",
    "Sales Executive",
    "Backend Developer",
    "HR Executive",
    "Financial Analyst",
    "Digital Marketer",
    "Operations Manager"
]

emp_roles = np.random.choice(roles_list, num_employees)

# Create DataFrame
data = pd.DataFrame({
    "emp_id": emp_ids,
    "emp_age": emp_ages,
    "emp_salary": emp_income,
    "role": emp_roles
})

# Save dataset
data.to_csv("staff_data.csv", index=False)
```

Q1

```
df = pd.read_csv("/content/staff_data.csv")

employ_age = np.array(df["emp_age"])
employ_salary = np.array(df["emp_salary"])
print(employ_age)
print(employ_salary)
print(f"employ_age data type {employ_age.dtype} and emp_salary data type {employ_salary.dtype}")

print(f"employ_age dimension {employ_age.ndim} and emp_salary dimension {employ_salary.ndim}")

print(f"employ_age shape {employ_age.shape} and emp_salary shape {employ_salary.shape}")

print(f"employ_age total number of elements {employ_age.size} and emp_salary total number of elements {employ_salary.size}")
```

Q2

```
maximum_salary = np.max(employ_salary)
minimum_salary = np.min(employ_salary)
average_salary = np.mean(employ_salary)
average_age = np.mean(employ_age)
print(f"maximum salary {maximum_salary}")
print(f"minimum salary {minimum_salary}")
print(f"average salary {average_salary}")
print(f"average age {average_age}")
```

Q3

```
age_filter = employ_age[(employ_age >= 25) & (employ_age <= 40)]
print(age_filter)
salary_filter = employ_salary[employ_salary > 50000]
print(salary_filter)
print(f"Total satisfy this conditon {age_filter.size + salary_filter.size}")
```

Q4

```
print(employ_age)
updated_age = np.where(employ_age >= average_age, employ_age + 2, employ_age + 4)
print(updated_age)
print(f"Average age for condition {np.mean(updated_age)}")
```

Q5

```
tech_roles_array = ["Software Developer", "Backend Developer", "Digital Marketer"]
print(employ_salary)
previous_total_salary = np.sum(employ_salary)
new_total_salary_expense = 0
for i, (job, salary) in enumerate(zip(roles_list, employ_salary)):
  if job in tech_roles_array:
    new_total_salary_expense = new_total_salary_expense + (salary * 0.08)
    employ_salary[i] = salary + (salary * 0.08)

print(f"new total salary expense {new_total_salary_expense}")
print(f"Old total {previous_total_salary}")
print(f"New total {np.sum(employ_salary)}")
```
