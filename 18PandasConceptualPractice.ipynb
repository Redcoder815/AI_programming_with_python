Mini project

```
data = {
    "Name": ["Alice", "Bob", "Charlie", "David", "Eva"],
    "Age": [20, 21, 19, 22, 20],
    "Marks": [85, 67, 90, 45, 88]
}

df = pd.DataFrame(data)
df
```

```
# add a grade columns

def assign_grade(marks):
  if marks >=80:
    return "A"
  elif marks >=60:
    return "B"
  else:
    return "C"


df["Grade"] = df["Marks"].apply(assign_grade)
```

```
df
```

```
# Filter Student

top_students = df[ df['Marks'] > 80 ]
top_students
```

```
sorted_df = df.sort_values( by="Marks", ascending= True )
sorted_df
```

```
#average marks

avg_marks = df["Marks"].mean()
avg_marks
```

```
df.to_csv("Student_output.csv",index=False)
```
