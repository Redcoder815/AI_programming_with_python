Dictionary

```
numbers = input().split()

freq = {}

for num in numbers:
  num = int(num)

  if num in freq:
    freq[num] += 1
  else:
    freq[num] = 1

print(freq)
```

Function

```
import numpy as np
import pandas as pd

def create_table():
  names = np.array(["Alice", "David", "Rakib", "Hasan"])
  score = np.array([90, 95, 99, 100])

  data = {
      "Name":names,
      "Score": score
  }

  df = pd.DataFrame(data)
  return df

table = create_table()
print(table)
```

Arguments

```
def total_sum(*args):
  sum = 0
  for num in args:
    sum = sum + num
  print(sum)

total_sum(10, 11, 23, 45)
```

```
def print_info(**kwarges):
  for key, value in kwarges.items():
    print(key, ":", value)

print_info(name="Rakib", age=22, city="Dhaka")
```

Iterator and Generator

```
df = pd.DataFrame({
    "A": [1, 2, 3],
    "B": [4, 5, 6]
})

# iterating row by row
for row in df.itertuples():
    print(row)   # each row is accessed one by one
```

```
# Generator is memory efficient (very important for large dataset in ML)

def data_loader(batch_size, data):
    # simulate large numpy dataset
    for i in range(0, len(data), batch_size):
        yield data[i:i+batch_size]   # yield returns chunk, not whole data

# Large dataset (numpy)
data = np.arange(20)   # [0,1,2,...,19]

loader = data_loader(5, data)

# get batches one by one
print(next(loader))  # first 5 elements
print(next(loader))  # next 5 elements

# ML insight:
# Instead of loading full dataset in memory,
# generator loads data in batches → used in training loop

# Example idea (conceptual):
# for batch in data_loader(32, large_dataset):
#     model.train(batch)
```

Filter function

```
df = pd.DataFrame({
    "Name": ["Alice", "Bob", "Rakib", "Hasan"],
    "Score": [90, 95, 98, 100]
})

filter_columns = df.filter(regex = "^S")
print(filter_columns)

high_score = df[df["Score"] > 95]
print(high_score)
```
