Number of lines, number of characters and number of words

```
from functools import reduce
with open("/content/sample_data/sample.txt", "r") as file:
  line_as_list = file.readlines()
  total_lines = len(line_as_list)
  print(total_lines)
  number_of_words = list(map(lambda x : len(x.split()), line_as_list))
  print(number_of_words)
  total_number_of_words = reduce(lambda x,y:x+y, number_of_words)
  print(total_number_of_words)

  remove_newline = list(map(str.strip, line_as_list))
  print(remove_newline)

  remove_space = list(map(lambda x:x.replace(" ",""), remove_newline))
  print(remove_space)

  character_list = list(map(lambda x : len(x), remove_space))

  total_number_of_characters = reduce(lambda x,y:x+y, character_list)

  print(total_number_of_characters)

with open("/content/sample_data/counter.txt", "w") as file:
  file.write(f"Total line {total_lines}\n Total words {total_number_of_words}\n Total characters {total_number_of_characters}")
```
