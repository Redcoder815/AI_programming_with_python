```
class Phone:
  def __init__(self, model, battery=100):
    self.model = model
    self.battery = battery
```

```
class SmartPhone(Phone):
  def __init(self, model, operating_system):
    super().__init__(model)
    self.operating_system = operating_system
```

```
class GamingPhone(SmartPhone):
  def __init__(self, model, operating_system, cooling_system):
    super().__init__(model, operating_system)
    self.cooling_system = cooling_system

  def start_game(self, name):
    print(f"Playing {name} on {self.model}")
```

```
model = GamingPhone("Samsung", "Mediatek", "Hydrogen")
```

```
model.start_game("Pes 26")
```

```
class MobilePhone:
  def __init__(self, model, battery=100):
    self.model = model
    self.__battery = battery

  def use(self):
    print("reduce battery")

  def charge(self):
    print("charge battery")

  def get_battery(self):
    print(f"Battery remains {self.__battery}")
```

```
new_model = MobilePhone("Huwaei", 97)
new_model.get_battery()
```

```
try:
  print(new_model.__battery)
except AttributeError:
  print("Can not access battery directly, because it is private")
```
