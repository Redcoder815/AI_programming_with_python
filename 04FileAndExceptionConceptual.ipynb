Lambda, map, filter, reduce function

```
greet = lambda: print("Hello everyone")

greet()
```

```
greet = lambda name, age: print("Hello ", name, "my age is ", age)

greet("Rakib", 24)
```

Map

```
arr = [1, 2, 3, 4, 5, 6]

def func1(x):
  return x * 20

arr2 = list(map(func1, arr))
print(arr2)
```

```
arr = [1, 2, 3, 4, 5, 6]

arr2 = list(map(lambda x: x*20, arr))

print(arr2)
```

```
score = [85, 90, 20, 100, 75, 65]

def grade(x):
  if x >= 80: return f"Score {x} is A+"
  if x >= 70: return f"Score {x} is A"
  if x >= 60: return f"Score {x} is A-"
  else: return f"Score {x} is F"

ans = list(map(grade, score))
print(ans)
```

Filter

```
score = [85, 90, 20, 100, 75, 65]

filter_data = list(filter(lambda x: x > 80, score))

print(filter_data)
```

```
score = [85, 90, 20, 100, 75, 65]

boolean_data = list(map(lambda x: x>80, score))

print(boolean_data)
```

```
from functools import reduce

score = [85, 90, 20, 100, 75, 65]

reduce(lambda x, y: x+y, score)
```

Using Context Manger = With

```
with open("sample.txt", "r") as f:
  print(f.read(5))
  print(f.read(5))
```

```
Big_data = ["Hello AI\n" for i in range(100)]

with open("sample1.txt", "w") as file:
  file.writelines(Big_data)

with open("sample1.txt", "r") as file:
  print(file.read())
```

```
with open("sample1.txt", "r") as file:
  chunk_size = 50

  chunk_data = file.read(chunk_size)

  while len(chunk_data) > 0:
    print(chunk_data)
    chunk_data = file.read(chunk_size)
```

Serialization and Deserialization
Serialization - process of converting python data types to JSON format
Deserialization - process of converting JSON to python data types

```
a = {
    "Name" : "Rakib",
    "Age" : 24
}
```

Serialization

```
import json

with open("sample3.txt", "w") as file:
  json.dump(a, file)
```

Deserialization

```
with open("sample3.txt", "r") as file:
  # print(json.load(file))
  print(type(json.load(file)))
```

raise Exception
Python raises exceptions automatically when something goes wrong.
You can also manually raise an exception using raise.
You can pass a message to explain why the exception happened.

```
raise NameError("Amar icca")
```

```
# --- DEMO: raising custom errors ---
def require_positive(n):
    if n <= 0:
        raise ValueError("n must be > 0")
    return n

def dividedbyn(n):
  if n==0:
    raise ZeroDivisionError("0 diye vag kora jabe nah")
  return 100/n;

try:
    # print(require_positive(-1))
    print(dividedbyn(0))
except ZeroDivisionError as e:
    print(e)
except ValueError as e:
    print("Caught:", e)
```
