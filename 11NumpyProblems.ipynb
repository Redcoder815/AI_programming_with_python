```
import numpy as np
```

```
celcius = [10, 15, 20, 25]

celcius_array = np.array(celcius)

farenheit = celcius_array * 1.8 + 32

print(farenheit)
```

```
seconds = [240, 200, 400, 600]

seconds_array = np.array(seconds)
print(seconds_array.size)
print(seconds_array.dtype)
print(seconds_array.shape)

minutes = seconds_array / 60

print(minutes)
```
