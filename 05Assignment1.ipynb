Given a list of integers, extract all numbers that are divisible by 3 but not divisible by 5.


```
numbers = input().split()
numbers = list(map(int, numbers))

numbers = list(filter(lambda x: x % 3 == 0, numbers))
numbers = list(filter(lambda x: x % 5 != 0, numbers))

print(numbers)
```

Write a function compute(a, b) that:


```
number_for_division = input().split()
number_for_division = list(map(int, number_for_division))

def compute(a, b):
  if (a - b) == 0:
    raise ZeroDivisionError("Error: Division not possible")
  return (a**2 + b**2) / (a - b)

try:
  print(compute(number_for_division[0], number_for_division[1]))
except ZeroDivisionError as e:
  print(e)
```

4 Write a program that converts a sentence into a list representing the length of each word.Ignore extra spaces


```
sentence = input().split()

length_of_words = list(map(lambda x: len(x), sentence))
print(length_of_words)
```

5 A number is called a Mirror Number if reversing its digits results in the same number.


```
number = int(input())
digit_list = []

while number > 0:
  digit = number % 10
  digit_list.append(digit)
  number = number // 10

reverse_digit_list = digit_list[::-1]

is_mirror = False

for i in range(len(digit_list)):
  if digit_list[i] == reverse_digit_list[i]:
    is_mirror = True
  else:
    is_mirror = False
    break

if is_mirror == True:
  print("Mirror Number")
else:
  print("Not Mirror Number")
```
