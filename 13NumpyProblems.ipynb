```
import numpy as np
```

Mean column and row

```
def calculate_mean(matrix, mode):
  mat = np.array(matrix)

  if mode == "row":
    means = np.mean(mat, axis = 1)
  else:
    means = np.mean(mat, axis = 0)

  return means

print(calculate_mean([[1,2,3,4], [5,6,7,8]], "row"))
print(calculate_mean([[1,2,3,4], [5,6,7,8]], "col"))
```

Matrix dot vector

```
def matrix_dot_vector(matrix, vector):
  mat = np.array(matrix)
  vec = np.array(vector)
  shape_a = mat.shape

  if shape_a[1] != vec.size:
    return -1
  return np.dot(mat, vec).tolist()

print(matrix_dot_vector([[1,2,3],[5,6,8]],[1,2,3]))
```

Create a 1D, 2D, and 3D NumPy array and print their dimensions and shapes.

```
one_d = np.array([1,2,3])
two_d = np.array([[1,2,3],[5,6,7]])
three_d = np.array([[[1,2,3],[5,6,7]]])

print(one_d.ndim)
print(two_d.ndim)
print(three_d.ndim)

print(one_d.shape)
print(two_d.shape)
print(three_d.shape)
```
