```
def create_file(filename, content):
  with open(filename, "w") as file:
    file.write(content)

create_file("note.txt", "Hi, I am Rakib hasan")
```

```
def read_lines(filename):
  with open(filename, "r") as file:
    content_as_list = file.readlines()
    # print(content_as_list)
    for i in range(0, len(content_as_list)):
      print(i, ":", content_as_list[i])

read_lines("note.txt")
```

```
def add_log(message):
  with open("log.txt", "a") as file:
    file.write(message + '\n')

add_log("User logged in")
add_log("File uploaded")
```

```
def read_from_position(filename, position):
  with open(filename, "r") as file:
    file.seek(position)
    print(file.read())

read_from_position("note.txt", 7)
```

```
def safe_read(filename):
  try:
    file = open(filename, "r")
  except Exception as e:
    print(e)
  else:
    print(file.read())
  finally:
    print("Operation complete")

safe_read("read.txt")
safe_read("log.txt")
```
