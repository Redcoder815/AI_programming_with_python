```
import pandas as pd
import numpy as np
```

Create a Pandas Series from the list [10, 20, 30, 40, 50]. Print the Series, its index, and data type. Then create a DataFrame with columns "Name", "Age", and "City" containing at least 4 rows of data.


```
data_list = [10, 20, 30, 40, 50]

df = pd.Series(data_list)
print(df)

print(np.array(df.index))
print(df.index)
print(df.dtype)
```

```
df_data = [
    {"Name": "Alice", "Age": 28, "City": "New York"},
    {"Name": "Bob", "Age": 34, "City": "Chicago"},
    {"Name": "Charlie", "Age": 22, "City": "San Francisco"},
    {"Name": "Diana", "Age": 31, "City": "Austin"}
]
df = pd.DataFrame(df_data)
print(df)
```



Using any DataFrame, write code to print the number of rows and columns, column data types (.dtypes), and summary statistics (.describe()).

```
print("Number of rows and columns ", df.shape)
print("column data types ", df.dtypes)
# print(df.dtypes)
print(df.describe())
```

```
data = {
    'Product': ['Pen', 'Book', 'Eraser', 'Pencil'],
    'Price': [10, 50, 5, 7],
    'Quantity': [100, 60, 200, 150]
}

df = pd.DataFrame(data)
print(df)
print("Print columns ", df.columns)
print("Index ", df.index)
print(df.iloc[1])
df['Price']
```

Using .loc[] and .iloc[], show the first row, the second column, and a 2×2 block    from the middle of your DataFrame.


```
print(df.loc[0])
print(df.iloc[0:1,:])
print(df.iloc[:,1])
#2x2 block from the middle of dataframe
print(df.iloc[1:3,0:2])
print(df.loc[1:2, ['Product', 'Price']])
```

```
df.drop(0, axis = 0)
df.drop('Price', axis = 1)
```
